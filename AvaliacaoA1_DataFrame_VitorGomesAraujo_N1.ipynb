#Reexecutando codigo para não acontecer erro de lógica e o dataframe nao remover o 5 registro(Julia) inves do index 6

```python
#Data frame de informações pessoais de uma rede social
import pandas as pd#Importação da biblioteca de manipulação de dados (Pandas)

colunas = ["Ano de entrada", "Usuario", "Senha", "Estado-Civil", "Sexo"]#Dados usados para columns
linhas = [1, 2, 3, 4, 5, 6]#Dados usados para linhas

ano_entrada = ["2024", "2014", "2015", "2018", "2019", "2021"]#Dados usados para dicionario
nome_de_usuario = ["Vitor", "Ana", "Roberto", "Romário", "Julia", "Gabriel"]#Dados usados para dicionario
senha = ["Imune4325", "@%#fawwan52", "awman214", "0reaisnacota", "linusgameplay5252165", "medizsuasenha2"]#Dados usados para dicionario
estado_civil = ["Solteiro/a", "Solteiro/a", "Solteiro/a", "Casado/a", "Solteiro/a", "Solteiro/a"]#Dados usados para dicionario
sexo = ["Masculino", "Feminino", "Masculino", "Masculino", "Feminino", "Masculino"]#Dados usados para dicionario

dados_users = {"Ano de entrada": ano_entrada, "Usuario": nome_de_usuario, "Senha": senha, "Estado-Civil": estado_civil, "Sexo": sexo}
#Criação do dicionario com as listas

df_users = pd.DataFrame(data = dados_users, columns = colunas, index = linhas)#Criação do dataframe
display(df_users)#Display do dataframe formatado
```

***
#6. Exclua um item (linha) de sua base de dados. Pontuação: 0,5 pontos
Você deve remover um item (linha) específico da base de dados

```python
df_users = df_users.drop(df_users.index[-1])
#Remove o ultimo index de registro do dataframe -1 representa a ultima posição como um vetor

print(df_users)#Print do dataframe
```

***
#7. Adicione um item (linha) na sua base de dados. Pontuação: 0,5 pontos
#Você deve adicionar um novo item (linha) à base de dados.

```python
novo_usuario = {"Ano de entrada": "2023",
                "Usuario": "Vitoria",
                "Senha": "ultracarro20",
                "Estado-Civil": "Solteiro/a",
                "Sexo": "Feminino"}
#Criando dados do novo usuario

df_users.loc[len(df_users) + 1] = novo_usuario
#Usando o len para ler a quantidade de registros e adicionando 1 para receber o novo usuario

print(df_users)#Print do resultado
```
