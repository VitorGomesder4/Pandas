```python
#Proposta 1: Dados de Clientes de uma Empresa de E-commerce
#a) Vendas Mensais
#b) Produto mais vendido
#c) Cliente que mais comprou
#d) Faturamento Mensal
#e) Adicionar uma nova coluna com o lucro
import pandas as pd

# Dados dos clientes
nomes_clientes = ["Viktor", "Gabriel", "Ben", "Paulo", "Ribamar"]
idades_clientes = [40, 28, 18, 19, 83]
cidade_clientes = ["Distrito Federal", "Recife", "Amapá", "Minas Gerais", "Distrito Federal"]

# Dados das vendas
vendas_mensais = [1000, 2000, 4700, 6420, 7000]
produtos_mais_vendidos = ["Celular", "Chip", "Placa-mãe", "Memória", "Processador"]
quantidade_comprada = [300, 220, 100, 520, 50]

# Crie um dicionário contendo os dados dos clientes
dados_clientes = {'Nome': nomes_clientes, 'Idade': idades_clientes, 'Cidade': cidade_clientes}

# Crie um DataFrame a partir do dicionário
clientes_df = pd.DataFrame(dados_clientes)
# Dados das vendas
dados_vendas = {
'Vendas Mensais': vendas_mensais,
'Produto mais vendido': produtos_mais_vendidos,
'Quantidade comprada': quantidade_comprada
}
# Crie um DataFrame a partir do dicionário
vendas_df = pd.DataFrame(dados_vendas)
# Adicionar uma nova coluna com o lucro
lucro = [2000, 212, 638, 2520, 3527]
vendas_df['Lucro'] = lucro
# a) Vendas Mensais
print("a) Vendas Mensais:")
print(vendas_df['Vendas Mensais'])
print()
# b) Produto mais vendido
produto_mais_vendido = vendas_df.loc[vendas_df['Vendas Mensais'].idxmax()]['Produto mais vendido']
print("b) Produto mais vendido:", produto_mais_vendido)
print()
# c) Cliente que mais comprou
cliente_mais_comprou = clientes_df.loc[vendas_df['Quantidade comprada'].idxmax()]['Nome']
print("c) Cliente que mais comprou:", cliente_mais_comprou)
print()
# d) Faturamento Mensal
faturamento_mensal = vendas_df['Vendas Mensais'].sum()
print("d) Faturamento Mensal:", faturamento_mensal)
print()
# e) Adicionar uma nova coluna com o lucro
print("e) DataFrame de Vendas com Lucro:")
print(vendas_df)
print()
```

```python
#Proposta 2: Dados de Estudantes de uma Escola
#a) Média de Notas
#b) Aluno com maior nota em Matemática
#c) Aluno mais novo
#d) Aluno com menor nota em História
#e) Adicionar uma nova coluna com a média geral das notas
import pandas as pd
# Dados dos alunos
nomes_alunos = ["Jonathan", "Danilo", "Juliana", "Gabriela", "Katrina"]
idades_alunos = [21, 20, 18, 17, 23]
notas_matematica = [99, 63, 53, 73, 80]
notas_historia = [45, 80, 90, 70, 80]
notas_portugues = [77, 78, 98, 75, 70]
# Crie um dicionário contendo os dados dos alunos
dados_alunos = {
'Nome': nomes_alunos,
'Idade': idades_alunos,
'Matemática': notas_matematica,
'História': notas_historia,
'Português': notas_portugues
}
# Crie um DataFrame a partir do dicionário
alunos_df = pd.DataFrame(dados_alunos)
# a) Média de Notas
print("a) Média de Notas:")
print(alunos_df[['Matemática', 'História', 'Português']].mean())
print()
# b) Aluno com maior nota em Matemática
aluno_maior_matematica = alunos_df.loc[alunos_df['Matemática'].idxmax()]['Nome']
print("b) Aluno com maior nota em Matemática:", aluno_maior_matematica)
print()
# c) Aluno mais novo
aluno_mais_novo = alunos_df.loc[alunos_df['Idade'].idxmin()]['Nome']
print("c) Aluno mais novo:", aluno_mais_novo)
print()
# d) Aluno com menor nota em História
aluno_menor_historia = alunos_df.loc[alunos_df['História'].idxmin()]['Nome']
print("d) Aluno com menor nota em História:", aluno_menor_historia)
print()
# e) Adicionar uma nova coluna com a média geral das notas
alunos_df['Média Geral'] = alunos_df[['Matemática', 'História', 'Português']].mean(axis=1)
print("e) DataFrame de Alunos com Média Geral:")
print(alunos_df)
print()
```

```python
#Proposta 3: Dados de Funcionários de uma Empresa
#a) Média de Salários
#b) Funcionário com maior salário
#c) Funcionário mais jovem
#d) Funcionário com menor salário
#e) Adicionar uma nova coluna com o total de horas extras
import pandas as pd
# Dados dos funcionários
nomes_funcionarios = ["Robert", "Flavio", "Pedri", "João", "Thomas"]
idades_funcionarios = [23, 52, 32, 19, 70]
salarios = [10000, 4256, 3000, 8000, 2000]
horas_extras = [4, 8, 2, 12, 7]
# Crie um dicionário contendo os dados dos funcionários
dados_funcionarios = {
'Nome': nomes_funcionarios,
'Idade': idades_funcionarios,
'Salário': salarios,
'Horas Extras': horas_extras
}
# Crie um DataFrame a partir do dicionário
funcionarios_df = pd.DataFrame(dados_funcionarios)
# a) Média de Salários
print("a) Média de Salários:")
print(funcionarios_df['Salário'].mean())
print()
#b) Funcionário com maior salário
print("b) Funcionário com maior salário:")
funcionario_maior_salario = funcionarios_df.loc[funcionarios_df['Salário'].idxmax()]
print(funcionario_maior_salario)
print()
# c) Funcionário mais jovem
print("c) Funcionário mais jovem:")
funcionario_mais_jovem = funcionarios_df.loc[funcionarios_df['Idade'].idxmin()]
print(funcionario_mais_jovem)
print()
# d) Funcionário com menor salário
print("d) Funcionário com menor salário:")
funcionario_menor_salario = funcionarios_df.loc[funcionarios_df['Salário'].idxmin()]
print(funcionario_menor_salario)
print()
# e) Adicionar uma nova coluna com o total de horas extras
funcionarios_df['Total de Horas Extras'] = funcionarios_df['Salário'] + (funcionarios_df['Horas Extras'] * 20) # Supondo que o valor da hora extra seja R$20
print("e) DataFrame de Funcionários com Total de Horas Extras:")
print(funcionarios_df)
print()
```

```python
#Proposta 4: Dados de Músicas de uma Plataforma de Streaming
#a) Mais ouvidas
#b) Artista com mais músicas
#c) Gênero mais popular
#d) Música mais recente
#e) Adicionar uma nova coluna com a duração em segundos
import pandas as pd
# Dados das músicas
nomes_musicas = ["Shape of You", "Bohemian Rhapsody", "Dance Monkey", "Blinding Lights", "Stairway to Heaven"]
artistas = ["Ed Sheeran", "Queen", "Tones and I", "The Weeknd", "Led Zeppelin"]
generos = ["Pop", "Rock", "Pop", "Indie", "Pop"]
ano_lancamento = [2017, 1975, 2019, 2019, 1971]
duracao_minutos = [3.54, 5.55, 3.30, 3.22, 8.02]
# Crie um dicionário contendo os dados das músicas
dados_musicas = {
'Música': nomes_musicas,
'Artista': artistas,
'Gênero': generos,
'Ano de Lançamento': ano_lancamento,
'Duração (minutos)': duracao_minutos
}
# Crie um DataFrame a partir do dicionário
musicas_df = pd.DataFrame(dados_musicas)
# a) Mais ouvidas
print("a) Mais ouvidas:")
mais_ouvida = musicas_df.loc[musicas_df['Ano de Lançamento'].idxmax()]['Música']
print(mais_ouvida)
print()
# b) Artista com mais músicas
contagem_artistas = musicas_df['Artista'].value_counts()
artista_mais_musicas = contagem_artistas.idxmax()
print("\nb) Artista com mais músicas:", artista_mais_musicas)
#c) Gênero mais popular
contagem_generos = musicas_df['Gênero'].value_counts()
genero_mais_popular = contagem_generos.idxmax()
print("\nc) Gênero mais popular:", genero_mais_popular)
#d) Música mais recente
musica_mais_recente = musicas_df.loc[musicas_df['Ano de Lançamento'].idxmax()]['Música']
print("\nd) Música mais recente:", musica_mais_recente)
#e) Adicionar uma nova coluna com a duração em segundos
musicas_df['Duração (segundos)'] = musicas_df['Duração (minutos)'] * 60
print("\ne) DataFrame de Músicas com Duração em Segundos:")
print(musicas_df)
print()
```

```python
#Proposta 5: Dados de Alimentação
#a) Calorias por porção
#b) Alimento com mais proteínas
#c) Alimento com menos calorias
#d) Alimento com mais calorias
#e) Adicionar uma nova coluna com a quantidade de carboidratos
import pandas as pd
# Dados dos alimentos
nomes_alimentos = ["Macarrão", "Salmão", "Brócolis", "Queijo", "Cenoura"]
porcao_g = [100, 100, 100, 50, 100]
calorias = [130, 200, 34, 400, 41]
proteinas_g = [5, 25, 2, 25, 1]
carboidratos_g = [25, 0, 6, 2, 10]
# Crie um dicionário contendo os dados dos alimentos
dados_alimentos = {
'Alimento': nomes_alimentos,
'Porção (g)': porcao_g,
'Calorias': calorias,
'Proteínas (g)': proteinas_g,
'Carboidratos (g)': carboidratos_g
}
# Crie um DataFrame a partir do dicionário
alimentos_df = pd.DataFrame(dados_alimentos)
print(alimentos_df)
print()

#a) Calorias por porção
alimentos_df['Calorias por porção'] = alimentos_df['Calorias'] / alimentos_df['Porção (g)']
print("a) Calorias por porção:")
print(alimentos_df[['Alimento', 'Calorias por porção']])
print()

#b) Alimento com mais proteínas
alimento_mais_proteinas = alimentos_df.loc[alimentos_df['Proteínas (g)'].idxmax()]['Alimento']
print("b) Alimento com mais proteínas:", alimento_mais_proteinas)
print()

#c) Alimento com menos calorias
alimento_menos_calorias = alimentos_df.loc[alimentos_df['Calorias'].idxmin()]['Alimento']
print("c) Alimento com menos calorias:", alimento_menos_calorias)
print()

#d) Alimento com mais calorias
alimento_mais_calorias = alimentos_df.loc[alimentos_df['Calorias'].idxmax()]['Alimento']
print("d) Alimento com mais calorias:", alimento_mais_calorias)
print()

#e) Adicionar uma nova coluna com a quantidade de carboidratos
alimentos_df['Carboidratos (g)'] = carboidratos_g
print("e) DataFrame de Alimentos com Quantidade de Carboidratos:")
print(alimentos_df)
print()
```
